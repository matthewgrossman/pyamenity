```python
amenity_blob = get_statistics(Amenities.WEIGHT_ROOM)
```

```python
def get_open_intervals(amenity_blob: Dict) -> Dict[Tuple[datetime, datetime], int]:
    ret = {}
    for interval_str, avail_str in amenity_blob['intervals'].items():
        if "Closed" in avail_str:
            continue
        taken, total = (int(n) for n in avail_str.split('/'))
        if taken >= total:
            continue
            
        interval = tuple(datetime.fromisoformat(s) for s in interval_str.split('/'))
        remaining_slots = total - taken
        ret[interval] = remaining_slots
        
    return ret
        
```

```python
open_intervals = get_open_intervals(amenity_blob)
```

```python
JSON({
    start.strftime("%m/%d: %I:%M-") + end.strftime("%I:%M%p"): s
    for (start, end), s in open_intervals.items()
})
```
